```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
thumbs_up_path = "/content/drive/MyDrive/gesture/thumbsup/"
thumbs_down_path = "/content/drive/MyDrive/gesture/thumbs_down_augmented_images/"

images = []
labels = []

for image_file in os.listdir(thumbs_up_path):
    img = cv2.imread(os.path.join(thumbs_up_path, image_file))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB
    img = cv2.resize(img, (224, 224))
    img = img / 255.0  # Normalize pixel values to [0, 1]
    images.append(img)
    labels.append(1)  # Thumbs-up label

for image_file in os.listdir(thumbs_down_path):
    img = cv2.imread(os.path.join(thumbs_down_path, image_file))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB
    img = cv2.resize(img, (224, 224))
    img = img / 255.0  # Normalize pixel values to [0, 1]
    images.append(img)
    labels.append(0)  # Thumbs-down label

images = np.array(images)
labels = np.array(labels)
```

```python
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)
```

```python

datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

datagen.fit(X_train)
```

```python
plt.figure(figsize=(10, 5))
for i in range(10):  # Display the first 10 images
    plt.subplot(2, 5, i + 1)
    plt.imshow(X_train[i])
    plt.title("Thumbs Up" if y_train[i] == 1 else "Thumbs Down")
    plt.axis('off')
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')  # Output layer for binary classification
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10)
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f'Test Accuracy: {test_accuracy}')
```
